## csv 파일을 이용하는 방법

```python
# csv파일을 열어 불러오기

import csv
f = open('subwayfee.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)

max_value = [0,0,0,0]
max_station = ['','','','']

for row in data : 
    for i in range(4,8):
        row[i] = int(row[i])
        if row[i] > max_value[i-4]:
            max_value[i-4] = row[i]
            max_station[i-4] = row[3]
print(max_value)
print(max_station)
            
        
```

```python
# csv파일을 열어 불러오기
# 0으로 나누어 발생한 에러

import csv
f = open('subwayfee.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)

max_rate = 0
rate = 0

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    
    rate = row[4] / row[6]
    if rate > max_rate:
        max_rate = rate
        print(row, rate)
```

```python
#0으로 나누지 않도록 수정하기

import csv
f = open('subwayfee.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)

max_rate = 0
rate = 0
station = ''

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    if row[6] != 0:
        rate = row[4] / row[6]
        if rate > max_rate:
            max_rate = rate
            print(row, rate)
```
